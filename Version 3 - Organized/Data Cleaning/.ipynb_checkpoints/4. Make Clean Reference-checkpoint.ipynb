```python
import pandas as pd
```

```python
FIRST_YEAR = 2000
CURRENT_YEAR = 2021
df = pd.read_csv(str(FIRST_YEAR) + '-' + str(CURRENT_YEAR-1) + ' Raw Data.csv', encoding='utf-8')
EMPTY_ROWS = [574]
```

```python
df
```

```python
# Add ID Column
df = df.reset_index()
df = df.rename(columns={'index':'ID'})

# Drop Unncessary Columns
df.drop(columns=['Round'], inplace = True)

# Convert Numerical
for i in df.columns[6:]:
    if 'DriveStart' in i and 'Rank' not in i:
        for j in df.index:
            if pd.isnull(df.loc[j, i]) == False:
                df.loc[j, i] = df.loc[j, i].split(' ')[1]
    elif 'DriveTime' in i and 'Rank' not in i:
        for j in df.index:
            if pd.isnull(df.loc[j, i]) == False:
                df.loc[j, i] = round(int(df.loc[j, i].split(':')[0])+int(df.loc[j, i].split(':')[1])/60, 2)
    else:
        df[i] = pd.to_numeric(df[i], downcast = 'float')

# Reorder Columns For Better Readability
listy = ['ID', 'Name', 'Position', 'Team', 'Pick', 'Year'] + list(df.columns)[6:]
df = df[listy]
df
```

```python
for i in EMPTY_ROWS:
    df.drop(i, inplace=True)
```

```python
df.to_csv(str(FIRST_YEAR) + '-' + str(CURRENT_YEAR-1) + ' Reference Data.csv', encoding='utf-8-sig', index=False)
```
