```python
# DataFrame
import pandas as pd
import numpy as np

# ML Models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier

# Train Test Split
from sklearn import model_selection

# Feature Scaling
from sklearn.preprocessing import StandardScaler

# Model Evaluation
from sklearn.metrics import classification_report, confusion_matrix

# Visualization
import matplotlib.pyplot as plt
```

```python
FIRST_YEAR = 2000
CURRENT_YEAR = 2021
```

```python
df = pd.read_csv(str(FIRST_YEAR) + '-' + str(CURRENT_YEAR) + ' Train Dataset.csv', encoding='utf-8')
df
```

```python
# Train Test Split
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(df.drop('Position', axis=1), df['Position'], 
                                                                    test_size = 0.2, stratify = df['Position'], random_state=42)
```

```python
X_train
```

```python
# Feature Scaling
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# For more see KNN Evaluation Notebook
# KNN Algorithm Evaluation With Neighbors (Lowest Error of .42)
knn = KNeighborsClassifier(n_neighbors=19)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Logistic Regression Evaluation
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Logistic Regression Highest Positive Correlations
coeff_df = pd.DataFrame(df.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False).head(10)
```

```python
# Logistic Regression Highest Negative Correlations
coeff_df.sort_values(by='Correlation', ascending=False).tail(10)
```

```python
# SVC Evaluation
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Gaussian Evaluation
gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Perceptron Evaluation
perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Linear SVC Evaluation
linear_svc = LinearSVC(max_iter=10000)
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# SGD Evaluation
sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Decision Tree Evaluation
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Random Forest Evaluation
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# See accuracy on training set
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
