```python
# POSITIONAL SPENDING PER TEAM
# https://overthecap.com/positional-spending/
URL = 'https://overthecap.com/positional-spending/'
response = get(URL)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find_all('table')
df = pd.read_html(str(table))[CURRENT_YEAR-2013]
df
```

```python
# TEAM SALARIES
# https://www.spotrac.com/nfl/los-angeles-chargers/cap/
BASE_URL = 'https://www.spotrac.com/nfl/'
TEAM = 'miami-dolphins'
```

```python
response = get(BASE_URL + TEAM + '/cap')
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
df
```

```python
df=df[[df.columns[0], 'Pos.', 'Cap Hit', 'Cap %']]
df['Cap Hit'] = [i for i in df['Cap Hit'].replace('[\$, ]', '', regex=True).astype(float)]
df
```

```python
df=df.dropna()
df
```

```python
df2 = df.sort_values('Cap Hit', ascending=False).drop_duplicates(['Pos.'])
df2
```
